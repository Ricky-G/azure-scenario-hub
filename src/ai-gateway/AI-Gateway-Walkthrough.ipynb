# AI Gateway Walkthrough

An end-to-end walkthrough of the **AI Gateway** built on Azure API Management in front of an existing Azure AI Foundry account. Each section is a working demo will run the cell to see the live behaviour, then read the explanation.

**What we'll cover**

1. The deployed environment
2. Two regional APIs over one shared backend pool
3. Managed-identity authentication to Foundry — no keys leave Azure
4. Per-application token chargeback with rich Application Insights dimensions
5. Per-product Tokens-Per-Minute throttling with clean 429 + Retry-After
6. Response caching for repeat requests
7. Edge request validation
8. Subscription-key authentication
9. Structured per-call tracing
10. Backend pool with circuit breakers
11. Graceful 5xx mock fallback
12. CORS for browser clients
13. Per-app cost estimation

> Each PowerShell cell is independent — you can run them in any order. Run the **setup cell** at the top once at the start; it sets the variables and helper functions every other cell depends on.

```polyglot-notebook
# Setup - run once at the start of the walkthrough.
# Set the resource group you deployed the AI Gateway scenario into.
$rg = 'rg-ai-gateway-demo'

# Set the resource group + account name of the existing Foundry.
$foundryRg      = '<foundry-resource-group>'
$foundryAccount = '<your-foundry-account>'

# Auto-discover the rest from the resource group.
$sub    = (az account show --query id -o tsv)
$apim   = (az resource list -g $rg --resource-type Microsoft.ApiManagement/service --query '[0].name' -o tsv)
$law    = (az resource list -g $rg --resource-type Microsoft.OperationalInsights/workspaces --query '[0].name' -o tsv)
$appi   = (az resource list -g $rg --resource-type Microsoft.Insights/components --query '[0].name' -o tsv)
$gw     = "https://$apim.azure-api.net"
$wsId   = (az monitor log-analytics workspace show -g $rg -n $law --query customerId -o tsv)
$appiId = (az resource show -g $rg -n $appi --resource-type Microsoft.Insights/components --query 'properties.AppId' -o tsv)

# Locate the scenario directory regardless of where VS Code launched the notebook.
$scenarioDir = $null
foreach ($candidate in @((Get-Location).Path, (Join-Path (Get-Location).Path 'src/ai-gateway'), $PSScriptRoot)) {
    if ($candidate -and (Test-Path (Join-Path $candidate 'bicep/main.bicep'))) { $scenarioDir = (Resolve-Path $candidate).Path; break }
}
if (-not $scenarioDir) { throw 'Could not locate the ai-gateway scenario directory. Set $scenarioDir manually.' }

function Get-Key([string]$product) {
    (az rest --method post `
        --url "https://management.azure.com/subscriptions/$sub/resourceGroups/$rg/providers/Microsoft.ApiManagement/service/$apim/subscriptions/$product-demo-sub/listSecrets?api-version=2023-05-01-preview" `
        -o json | ConvertFrom-Json).primaryKey
}

function Get-Header($response, [string]$name) {
    $v = $response.Headers[$name]
    if ($null -eq $v) { return $null }
    if ($v -is [System.Collections.IEnumerable] -and $v -isnot [string]) { return ($v | Select-Object -First 1) }
    return $v
}

function Invoke-AppInsightsKql([string]$Query) {
    $token = az account get-access-token --resource https://api.applicationinsights.io --query accessToken -o tsv
    $body  = @{ query = $Query } | ConvertTo-Json -Compress
    $r = Invoke-RestMethod -Method Post `
        -Uri "https://api.applicationinsights.io/v1/apps/$appiId/query" `
        -Headers @{ Authorization = "Bearer $token"; 'Content-Type' = 'application/json' } `
        -Body $body
    if ($r.tables[0].rows.Count -eq 0) { Write-Host '(no rows yet - wait 60-90s for telemetry ingestion)' -ForegroundColor Yellow; return }
    $cols = $r.tables[0].columns.name
    $r.tables[0].rows | ForEach-Object {
        $row = $_; $obj = [ordered]@{}
        for ($i = 0; $i -lt $cols.Count; $i++) { $obj[$cols[$i]] = $row[$i] }
        [pscustomobject]$obj
    } | Format-Table -AutoSize
}

function Invoke-LawKql([string]$Query) {
    $token = az account get-access-token --resource https://api.loganalytics.io --query accessToken -o tsv
    $body  = @{ query = $Query } | ConvertTo-Json -Compress
    $r = Invoke-RestMethod -Method Post `
        -Uri "https://api.loganalytics.io/v1/workspaces/$wsId/query" `
        -Headers @{ Authorization = "Bearer $token"; 'Content-Type' = 'application/json' } `
        -Body $body
    if ($r.tables[0].rows.Count -eq 0) { Write-Host '(no rows yet - wait 60-90s for telemetry ingestion)' -ForegroundColor Yellow; return }
    $cols = $r.tables[0].columns.name
    $r.tables[0].rows | ForEach-Object {
        $row = $_; $obj = [ordered]@{}
        for ($i = 0; $i -lt $cols.Count; $i++) { $obj[$cols[$i]] = $row[$i] }
        [pscustomobject]$obj
    } | Format-Table -AutoSize
}

Write-Host '==> Setup complete' -ForegroundColor Green
Write-Host "    APIM gateway   : $gw"
Write-Host "    LAW workspace  : $law"
Write-Host "    App Insights   : $appi"
Write-Host "    Scenario dir   : $scenarioDir"
```

## 2. Two regional APIs over one backend pool

The gateway exposes the same Foundry backend through two separate APIs:

- `Azure OpenAI - Australia East` at `/aue/openai/...`
- `Azure OpenAI - Global` at `/global/openai/...`

Same subscription key works against either. Each response carries `x-ai-gateway-route` and `x-ai-gateway-region` headers so the caller can prove which API served the request — even when the response was served from cache.

```polyglot-notebook
$key  = Get-Key 'retail-smart-shopping'
$body = @{
    messages   = @(@{ role = 'user'; content = 'Reply with the single word OK.' })
    max_tokens = 5
} | ConvertTo-Json -Compress

foreach ($route in 'aue', 'global') {
    $r = Invoke-WebRequest -Method Post `
        -Uri "$gw/$route/openai/deployments/gpt-4.1-mini/chat/completions?api-version=2024-10-21" `
        -Headers @{ 'api-key' = $key; 'Content-Type' = 'application/json' } `
        -Body $body -SkipHttpErrorCheck
    [pscustomobject]@{
        Route   = $route
        Status  = $r.StatusCode
        XRoute  = (Get-Header $r 'x-ai-gateway-route')
        XRegion = (Get-Header $r 'x-ai-gateway-region')
        XCache  = (Get-Header $r 'x-ai-gateway-cache')
    }
}
```

```polyglot-notebook
# Confirm the backend pool composition: priority-1 primary, priority-2 secondary.
az rest --method get `
    --url "https://management.azure.com/subscriptions/$sub/resourceGroups/$rg/providers/Microsoft.ApiManagement/service/$apim/backends/openai-backend-pool?api-version=2024-05-01" `
    --query '{type:properties.type, members:properties.pool.services[].{id:id, priority:priority, weight:weight}}' -o json
```

## 3. Managed-identity authentication to Foundry

The caller authenticates to APIM with a subscription key. APIM then strips that header and authenticates to Foundry using its own **system-assigned managed identity**. The Foundry account holds no shared secrets — only an Azure RBAC role assignment (`Cognitive Services User`) for the APIM principal.

```polyglot-notebook
$key  = Get-Key 'retail-smart-shopping'
$body = @{
    messages   = @(@{ role = 'user'; content = "MI test - reply 'managed identity'" })
    max_tokens = 10
} | ConvertTo-Json -Compress

$r = Invoke-WebRequest -Method Post `
    -Uri "$gw/global/openai/deployments/gpt-4.1-mini/chat/completions?api-version=2024-10-21" `
    -Headers @{ 'api-key' = $key; 'Content-Type' = 'application/json' } `
    -Body $body -SkipHttpErrorCheck
$payload = $r.Content | ConvertFrom-Json
[pscustomobject]@{
    Status = $r.StatusCode
    Reply  = $payload.choices[0].message.content
}
```

```polyglot-notebook
# Show the role assignment proving APIM's MI has access to the Foundry account.
$apimMi  = az apim show -g $rg -n $apim --query identity.principalId -o tsv
$foundry = az cognitiveservices account show -g <foundry-resource-group> -n <your-foundry-account> --query id -o tsv
az role assignment list --scope $foundry --assignee $apimMi `
    --query '[].{role:roleDefinitionName, principalType:principalType, scope:scope}' -o table
```

## 4. Per-application token chargeback

Every API call is tagged at the gateway with `App ID`, `Use Case`, `Product Name`, `Deployment`, `Route`, and `Region Label`. The gateway emits a custom Application Insights metric per call, and a structured trace event with the full set of dimensions, so finance can carve consumption by cost centre.

The next cell drives realistic traffic across both APIs. Then we wait ~60–90 seconds for telemetry ingestion and run a KQL query against the resulting `customMetrics` table.

```polyglot-notebook
Push-Location $scenarioDir
try {
    ./test-harness/Invoke-Demo.ps1 `
        -ResourceGroupName $rg `
        -ApimName $apim `
        -CallsPerProduct 2 `
        -Route both
}
finally { Pop-Location }
```

```polyglot-notebook
# Wait briefly for App Insights ingestion, then run the chargeback query.
Write-Host 'Waiting 75s for App Insights ingestion...' -ForegroundColor DarkGray
Start-Sleep -Seconds 75

Invoke-AppInsightsKql @"
customMetrics
| where timestamp > ago(15m)
| where name == 'Total Tokens'
| extend appId      = tostring(customDimensions['App ID'])
| extend useCase    = tostring(customDimensions['Use Case'])
| extend product    = tostring(customDimensions['Product Name'])
| extend route      = tostring(customDimensions['Route'])
| extend region     = tostring(customDimensions['Region Label'])
| extend deployment = tostring(customDimensions['Deployment'])
| summarize Calls = count(), Tokens = sum(value) by appId, useCase, product, route, region, deployment
| order by Tokens desc
"@
```

## 5. Per-product Tokens-Per-Minute throttling

Each product carries an independent TPM ceiling. `finance-smart-analysis` (5 000 TPM) trips first; `customer-care-chat` (10 000 TPM) and `retail-smart-shopping` (20 000 TPM) keep flowing. APIM returns a clean `429` with a `Retry-After` header, and the policy emits a structured warning trace per throttled call so SREs can alert on it.

```polyglot-notebook
Push-Location $scenarioDir
try {
    ./test-harness/Invoke-Throttle.ps1 `
        -ResourceGroupName $rg `
        -ApimName $apim `
        -Product finance-smart-analysis `
        -Calls 12 `
        -Route global
}
finally { Pop-Location }
```

```polyglot-notebook
# Wait briefly then query the throttling-event traces.
Write-Host 'Waiting 60s for ingestion...' -ForegroundColor DarkGray
Start-Sleep -Seconds 60

Invoke-AppInsightsKql @"
traces
| where timestamp > ago(15m)
| where message == 'ai-gateway-throttling-event'
| extend appId      = tostring(customDimensions.appId)
| extend product    = tostring(customDimensions.productName)
| extend retryAfter = toint(customDimensions.retryAfter)
| summarize Throttled = count(), Min_RetryAfter = min(retryAfter), Max_RetryAfter = max(retryAfter) by appId, product
"@
```

## 6. Response cache for repeat requests

The gateway holds successful non-streaming responses in APIM's internal cache for 5 minutes, keyed by a SHA-256 hash of `(deployment + product + body)`. Identical requests return in well under a second and cost zero model tokens. Each response carries `x-ai-gateway-cache: HIT` or `MISS` so the caller knows where the bytes came from.

```polyglot-notebook
$key   = Get-Key 'retail-smart-shopping'
$stamp = Get-Date -Format 'yyyyMMddHHmmssfff'
$body  = @{
    messages    = @(@{ role = 'user'; content = "Cache demo $stamp - reply with the single word OK." })
    max_tokens  = 5
    temperature = 0.0
} | ConvertTo-Json -Compress

1..3 | ForEach-Object {
    $sw = [System.Diagnostics.Stopwatch]::StartNew()
    $r = Invoke-WebRequest -Method Post `
        -Uri "$gw/aue/openai/deployments/gpt-4.1-mini/chat/completions?api-version=2024-10-21" `
        -Headers @{ 'api-key' = $key; 'Content-Type' = 'application/json' } `
        -Body $body -SkipHttpErrorCheck
    $sw.Stop()
    [pscustomobject]@{
        Call       = $_
        Status     = $r.StatusCode
        Cache      = (Get-Header $r 'x-ai-gateway-cache')
        DurationMs = $sw.ElapsedMilliseconds
    }
}
```

## 7. Edge request validation

Malformed requests are rejected at the gateway before they can burn model capacity. For chat completions the gateway requires a non-empty `messages` array; anything else gets a `400` with an `ai_gateway_invalid_request` error code. The Foundry backend never sees the call.

```polyglot-notebook
$key = Get-Key 'retail-smart-shopping'
$bad = @{ prompt = 'hello' } | ConvertTo-Json -Compress  # intentionally missing 'messages'

$r = Invoke-WebRequest -Method Post `
    -Uri "$gw/global/openai/deployments/gpt-4.1-mini/chat/completions?api-version=2024-10-21" `
    -Headers @{ 'api-key' = $key; 'Content-Type' = 'application/json' } `
    -Body $bad -SkipHttpErrorCheck

Write-Host "Status : $($r.StatusCode)"
Write-Host "Body   : $($r.Content)"
```

## 8. Subscription-key authentication

Every call is gated by an APIM subscription. A request without a valid key gets a `401` from APIM — Foundry never sees the call. Key rotation, suspension, per-key analytics and quota are all handled by APIM Subscriptions out of the box.

```polyglot-notebook
$body = @{
    messages   = @(@{ role = 'user'; content = 'hi' })
    max_tokens = 5
} | ConvertTo-Json -Compress

$r = Invoke-WebRequest -Method Post `
    -Uri "$gw/global/openai/deployments/gpt-4.1-mini/chat/completions?api-version=2024-10-21" `
    -Headers @{ 'api-key' = 'not-a-real-key'; 'Content-Type' = 'application/json' } `
    -Body $body -SkipHttpErrorCheck

Write-Host "Status : $($r.StatusCode)"
Write-Host "Body   : $($r.Content)"
```

## 9. Structured per-call tracing

On every successful non-streaming response the gateway emits a `<trace>` event into Application Insights with the full set of dimensions: app, use case, product, route, region, deployment, model, prompt/completion/total tokens, session id, end-user id, and gateway region. Auditors and SREs can pivot on any column.

```polyglot-notebook
Invoke-AppInsightsKql @"
traces
| where timestamp > ago(20m)
| where message == 'ai-gateway-token-usage'
| extend appId       = tostring(customDimensions.appId)
| extend route       = tostring(customDimensions.apiRoute)
| extend region      = tostring(customDimensions.apiRegion)
| extend totalTokens = toint(customDimensions.totalTokens)
| extend sessionId   = tostring(customDimensions.sessionId)
| extend endUserId   = tostring(customDimensions.endUserId)
| project timestamp, appId, route, region, totalTokens, sessionId, endUserId
| order by timestamp desc
| take 8
"@
```

## 10. Backend pool with circuit breakers

Both APIs route through the **`openai-backend-pool`** rather than a single backend. The pool has a primary (priority 1) and a secondary (priority 2). Each member has a circuit breaker that trips on 5 server errors in a 1-minute window and stays open for 30 seconds. APIM transparently promotes priority-2 traffic during the trip window; in production the secondary backend would point at a paired-region Foundry for true active/passive failover.

```polyglot-notebook
Write-Host '== Pool composition ==' -ForegroundColor Cyan
az rest --method get `
    --url "https://management.azure.com/subscriptions/$sub/resourceGroups/$rg/providers/Microsoft.ApiManagement/service/$apim/backends/openai-backend-pool?api-version=2024-05-01" `
    --query '{type:properties.type, members:properties.pool.services[].{priority:priority, weight:weight}}' -o json

Write-Host "`n== Primary backend circuit breaker ==" -ForegroundColor Cyan
az rest --method get `
    --url "https://management.azure.com/subscriptions/$sub/resourceGroups/$rg/providers/Microsoft.ApiManagement/service/$apim/backends/openai-backend-primary?api-version=2024-05-01" `
    --query 'properties.circuitBreaker' -o json
```

## 12. CORS for browser clients

A service-level CORS policy lives on the APIM instance, so a browser SPA can call the gateway directly without a thin proxy. The `expose-headers` list includes the AI-Gateway-specific headers (`x-ai-gateway-cache`, `-route`, `-fallback`) so JavaScript can read them.

```polyglot-notebook
$r = Invoke-WebRequest -Method Options `
    -Uri "$gw/global/openai/deployments/gpt-4.1-mini/chat/completions" `
    -Headers @{
        'Origin'                         = 'https://contoso.example'
        'Access-Control-Request-Method'  = 'POST'
        'Access-Control-Request-Headers' = 'content-type,api-key'
    } -SkipHttpErrorCheck

[pscustomobject]@{
    Status       = $r.StatusCode
    AllowOrigin  = (Get-Header $r 'Access-Control-Allow-Origin')
    AllowMethods = (Get-Header $r 'Access-Control-Allow-Methods')
    AllowHeaders = (Get-Header $r 'Access-Control-Allow-Headers')
}
```

## 13. Per-app cost estimate

With token counts and rich dimensions in `customMetrics`, mapping consumption to dollars is a single KQL query. The `pricing` datatable below uses publicly listed `gpt-4.1` and `gpt-4.1-mini` rates per 1K tokens — replace with your enterprise contract rates before sharing externally.

```polyglot-notebook
Invoke-AppInsightsKql @'
let pricing = datatable(model:string, pricePerKPrompt:real, pricePerKCompletion:real)
[
    "gpt-4.1",      0.0030, 0.0120,
    "gpt-4.1-mini", 0.0004, 0.0016,
    "unknown",      0.0010, 0.0040
];
customMetrics
| where timestamp > ago(24h)
| where name in ("Prompt Tokens", "Completion Tokens")
| extend appId      = tostring(customDimensions["App ID"])
| extend route      = tostring(customDimensions["Route"])
| extend deployment = tostring(customDimensions["Deployment"])
| where isnotempty(route)
| summarize tokens = sum(value) by appId, route, deployment, name
| evaluate pivot(name, sum(tokens), appId, route, deployment)
| extend PromptTokens     = coalesce(toint(["Prompt Tokens"]),     0)
| extend CompletionTokens = coalesce(toint(["Completion Tokens"]), 0)
| join kind=leftouter pricing on $left.deployment == $right.model
| extend EstCostUSD = round(
        (PromptTokens     / 1000.0) * coalesce(pricePerKPrompt,     0.001)
      + (CompletionTokens / 1000.0) * coalesce(pricePerKCompletion, 0.004), 4)
| project appId, route, deployment, PromptTokens, CompletionTokens, EstCostUSD
| order by EstCostUSD desc
'@
```

## Wrap-up

- **Two clearly-labelled regional APIs** sharing one backend pool.

- **Per-product TPM throttling** with structured 429 + Retry-After.
- **Response caching** delivering ~2× faster responses for repeat traffic at zero token cost.
- **Edge request validation** that rejects malformed traffic before it reaches the model.
- **Subscription-key gating** with first-class APIM subscription management.
- **Structured per-call tracing** for audit and debugging.
- **Backend pool with circuit breakers** for production-grade resilience.
- **Graceful 5xx mock fallback** so client apps degrade cleanly.
- **CORS** for browser-based clients.
- **Cost estimation** mapping tokens to dollars per app and route.
